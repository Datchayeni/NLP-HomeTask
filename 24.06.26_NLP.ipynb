```python
from collections import Counter
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')

# Sample News Articles
articles = """
Artificial intelligence is transforming healthcare and education systems.
AI and machine learning are the future of technology and innovation.
Healthcare systems are improving with artificial intelligence solutions.
Education technology is growing rapidly with AI based tools.
"""

# 1. Convert text to lowercase
text = articles.lower()

# 2. Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# 3. Tokenization
words = word_tokenize(text)

# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

print("Filtered Tokens:")
print(filtered_words)

# 5. Word Frequency Distribution
freq_dist = Counter(filtered_words)

print("\nWord Frequency Distribution:")
for word, count in freq_dist.items():
    print(word, ":", count)

# 6. Vocabulary (Unique Words)
vocabulary = set(filtered_words)

print("\nVocabulary:")
print(vocabulary)

# 7. Top 5 Most Frequent Words
top5 = freq_dist.most_common(5)

print("\nTop 5 Frequent Words:")
for word, count in top5:
    print(word, ":", count)

# 8. Type-Token Ratio (TTR)
total_words = len(filtered_words)
unique_words = len(vocabulary)

ttr = unique_words / total_words

print("\nTotal Words =", total_words)
print("Unique Words =", unique_words)
print("Type-Token Ratio (TTR) =", round(ttr, 2))

# 9. Dominant Topic Analysis
print("\nDominant Topic Analysis:")
print("The most frequent words are related to AI, healthcare, education, and technology.")
print("Hence, the dominant topic is Artificial Intelligence and Technology.")

# 10. Bar Chart Visualization
words_chart = [word for word, count in top5]
counts_chart = [count for word, count in top5]

plt.figure(figsize=(8,5))
plt.bar(words_chart, counts_chart)
plt.title("Top 5 Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```
