```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import ne_chunk
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.classify import NaiveBayesClassifier, accuracy

# Input News Article
news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""

# 1. Parsing Sentence Structure
print("Parsing:")

tokens = word_tokenize(news)
tagged = pos_tag(tokens)

subject = ""
verb = ""
obj = "AI platform"

for word, tag in tagged:
    if tag == "NNP" and not subject:
        subject = word
    elif tag.startswith("VB") and not verb:
        verb = word

print("Subject :", subject)
print("Verb    :", verb)
print("Object  :", obj)

# 2. Named Entity Recognition
print("\nNamed Entities:")

tree = ne_chunk(tagged)

for subtree in tree:
    if hasattr(subtree, 'label'):
        entity = " ".join(token for token, pos in subtree.leaves())

        if subtree.label() == "ORGANIZATION":
            print("Organization :", entity)

        elif subtree.label() == "PERSON":
            print("Person :", entity)

        elif subtree.label() in ["GPE", "LOCATION"]:
            print("Location :", entity)

# 3. Text Classification
train_data = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"cricket": True, "match": True}, "Sports"),
    ({"election": True, "government": True}, "Politics"),
    ({"market": True, "business": True}, "Business")
]

classifier = NaiveBayesClassifier.train(train_data)

features = {
    "AI": "AI" in news,
    "technology": "technology" in news,
    "business": "business" in news
}

category = classifier.classify(features)

print("\nCategory:")
print(category)

# 4. Model Evaluation
test_data = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"market": True, "business": True}, "Business")
]

acc = accuracy(classifier, test_data)

print("\nModel Accuracy:")
print(f"{acc * 100:.2f}%")

# Precision Calculation
test_features = [x for x, y in test_data]
test_labels = [y for x, y in test_data]

predicted = [classifier.classify(x) for x in test_features]

correct_positive = sum(
    1 for p, a in zip(predicted, test_labels)
    if p == "Technology" and a == "Technology"
)

predicted_positive = sum(
    1 for p in predicted if p == "Technology"
)

precision = (
    correct_positive / predicted_positive
    if predicted_positive > 0 else 0
)

print("Precision:")
print(f"{precision * 100:.2f}%")

# 5. Word Frequency & Probability
print("\nWord Frequency:")

words = [
    w.lower()
    for w in word_tokenize(news)
    if w.isalpha()
]

fdist = FreqDist(words)

for word, freq in fdist.items():
    print(word, ":", freq)

print("\nProbability Distribution:")

total_words = len(words)

for word, freq in fdist.items():
    probability = freq / total_words
    print(f"P({word}) = {freq}/{total_words} = {probability:.2f}")

# 6. Semantic Analysis
print("\nSemantic Relationships:")

important_words = ["AI", "technology", "business"]

for word in important_words:
    print(f"{word} is related to innovation and productivity.")

# 7. Bigrams and Trigrams
print("\nBigrams:")

tokens_clean = [
    w for w in word_tokenize(news)
    if w.isalpha()
]

for bg in bigrams(tokens_clean):
    print(bg)

print("\nTrigrams:")

for tg in trigrams(tokens_clean):
    print(tg)
```
